2. b) Parser

```python
from nltk import CFG
```

```python
grammar = CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> 'the' N | N PP | 'the' N PP
VP -> V NP | V PP | V NP PP
N -> 'cat'
N -> 'dog'
N -> 'rug'
V -> 'chased'
V -> 'sat'
P -> 'in'
P -> 'on'
""")
```

```python
from nltk.parse import RecursiveDescentParser
from nltk.parse import ShiftReduceParser
rd = RecursiveDescentParser(grammar)
sr = ShiftReduceParser(grammar)
```

```python

sentence1 = 'the cat chased the dog'.split()

sentence2 = 'the cat chased the dog on the rug'.split()
```

```python
def rdparseSen(sentence):
    for t in rd.parse(sentence):
        print(t)
rdparseSen(sentence1)
rdparseSen(sentence2)
```

```python
def srparseSen(sentence):
    for t in sr.parse(sentence):
        print(t)
srparseSen(sentence1)
srparseSen(sentence2)
```

3. Create a text corpus with a minimum of 200 words (unique content). Implement the
following text processing

```python
text="Ancient manuscripts also divided sentences into paragraphs with line breaks (newline) followed by an initial at the beginning of the next paragraph. An initial is an oversized capital letter, sometimes outdented beyond the margin of the text. This style can be seen, for example, in the original Old English manuscript of Beowulf. Outdenting is still used in English typography, though not commonly. Modern English typography usually indicates a new paragraph by indenting the first line. This style can be seen in the (handwritten) United States Constitution from 1787. For additional ornamentation, a hedera leaf or other symbol can be added to the inter-paragraph white space, or put in the indentation space.On a typewriter, a double carriage return produces a blank line for this purpose; professional typesetters may put in an arbitrary vertical space by adjusting leading. This style is very common in electronic formats, such as on the World Wide Web and email. Wikipedia itself employs this format."
```

```python
def textLower(text):
    return text.lower()
```

```python
text=textLower(text)
```

```python
from nltk.tokenize import word_tokenize
 
tokens = word_tokenize(text)
len(tokens)
```

```python
sentences = nltk.sent_tokenize(text)  
for sentence in sentences:
	print(sentence)
	print("\n")
```

```python

# from nltk.corpus import stopwords
without_sw=[word for word in tokens if not word in stopwords.words()]
len(without_sw)
```

```python

stem_words=[]
ps = PorterStemmer()
for w in without_sw:
    stem_words.append(ps.stem(w))
len(stem_words)
```

```python
lem_words=[]
lemmatizer = WordNetLemmatizer()
for w in stem_words:
    lem_words.append(lemmatizer.lemmatize(w))
len(lem_words)
```

```python
tagged = nltk.pos_tag(lem_words)
tagged
```
